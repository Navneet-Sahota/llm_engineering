```python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants
# Same idea as day5: run everything locally through Ollama (no key, no cost).
# Swap MODEL_GPT to 'gpt-4o-mini' and use a real OpenAI() client if you have a key + credits.

MODEL_LLAMA = 'qwen2.5:7b'    # "frontier-style" answer — non-thinking, fast (you have this pulled)
```

```python
# set up environment

load_dotenv(override=True)

# Same pattern as day5: talk to local Ollama through the OpenAI client
openai = OpenAI(base_url=os.getenv('OLLAMA_BASE_URL'), api_key="ollama")
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get the frontier model to answer, with streaming (typewriter effect, like day5)

def ask(question, model):
    stream = openai.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": question}],
        stream=True,
    )
    response = ""
    handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        update_display(Markdown(response), display_id=handle.display_id)

ask(question, MODEL_LLAMA)
```
